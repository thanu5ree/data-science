clearing the outliers

```python
q1=df['MonthlyCharge'].quantile(0.25)
print(q1)
```

```python
cat.boxplot(y='MonthlyCharge',data=df)
```

```python
q2=df['MonthlyCharge'].quantile(0.75)
print(q2)
```

```python
iqr=q2-q1
```

fixing lower limit and upper limit

```python
df[(df['MonthlyCharge']>=q1-1.5*iqr)&(df['MonthlyCharge']<=q2+1.5*iqr)]
```

```python
df_new=df[(df['MonthlyCharge']>=q1-1.5*iqr) & (df['MonthlyCharge']<=q2+1.5*iqr)]
```

```python
df_new
```

```python
cat.boxplot( y='MonthlyCharge', data=df_new)
```

```python
cat.boxplot(x='CustServCalls', y='MonthlyCharge', data=df_new)
```

almost of the outlier is dissappeared successfully

```python
cat.boxplot(x='CustServCalls', y='OverageFee', data=df_new)
```

```python
cat.boxplot(y='CustServCalls', x='Churn', data=df_new)
```

```python
cat.scatterplot(y='MonthlyCharge', x='OverageFee', hue='Churn', data=df_new)
```
